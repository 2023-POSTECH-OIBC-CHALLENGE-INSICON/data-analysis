```python
import warnings
warnings.filterwarnings('ignore')
import os
os.getcwd()
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',100)
```

```python
import pandas as pd

model_df = pd.read_csv('./models.csv')
model_df.head()
```

```python
weather_forecast_df = pd.read_csv('./OIBC2023_data/weather_forecast.csv')
```

```python
def diff_correlation(df):
    df = df[df["amount"] != 0]
    df['diff'] = ((df['amount'] - df['shot']) / df['amount']) * 100
    df["diff"] = df["diff"].abs()
    diff_df = df[["time", "diff"]]
    weather_diff_df = pd.merge(diff_df, weather_forecast_df, on='time', how='inner')
    numeric_cols = weather_diff_df.select_dtypes(include=[np.number])
    correlations = numeric_cols.corrwith(numeric_cols['diff']).abs()
    correlations = numeric_cols.corrwith(numeric_cols['diff'])
    print(correlations.sort_values())
```

## model 별 round별

```python
for i in range(0, 5):
    m_df =  model_df[["time", "round", f"model_id_{i}", "amount"]]
    for j in range(1, 3):
        print(f"\n\n\n\nmodel {i} round {j}")
        mr_df = m_df[m_df["round"] == j]
        mr_df.drop("round", axis=1, inplace=True)
        mr_df.rename(columns = {f"model_id_{i}": "shot"}, inplace=True)
        diff_correlation(mr_df)
```
